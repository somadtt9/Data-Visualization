```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
my_data = pd.read_csv(r"C:\Users\SOMA\Downloads\Pokemon1 (1).csv", index_col=0, header=0)
```

```python
my_data
```

```python
# scatter plot is used to visualize the relationship between 2 continuous numeric variables.

plt.scatter(data= my_data, x = 'Attack', y = 'Defense')
```

```python
plt.figure(figsize=(20,10))
plt.scatter(data=my_data, x= 'Attack',y = 'Defense', s =500, c = 'green', marker = '*')


#Change axes range
plt.xlim(0,150)
plt.ylim(0,200)
#add title
plt.title("Relationship between Attack and Defense", size = 20)
#add x and y labels
plt.xlabel('Attack Points', size=15)
plt.ylabel('Defense Points', size=15)
plt.grid(True)
#show plot
plt.savefig('test.jpg')
plt.show()
```

```python
#scatter plot with regression line

sns.lmplot(data= my_data, x = 'Attack', y = 'Defense')
```

```python
sns.lmplot(data= my_data, x = 'Attack', y= 'Defense', fit_reg = False, hue= 'Stage',palette = 'Set1', markers = ['*','^','o'])
```

```python
sns.lmplot(data= my_data, x = 'Attack', y= 'Defense', fit_reg = False, hue= 'Legendary',palette = 'Set1', markers = ['*','^'])
```

```python
# pairplot: when we need more than one graph in a single output for a particular dataframe we use pair plot.

sns.pairplot(data=my_data, x_vars = ['Attack','Defense','Speed'],
            y_vars = ['Total','HP'],
            kind = 'scatter')
```

```python
# Histogram :- Histogram are useful to visualize the frequency distribution of single continuous variable.
import numpy as np
plt.hist(x = 'Attack', data = my_data, bins = 10, color = 'darkblue')
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(40,10))
plt.subplot(131)  ## 1 row 3 column 1st position
plt.hist(x = 'Attack', data = my_data, bins = 10, color = 'darkblue')
plt.title('Attacks')

plt.subplot(132)  ##1 row 3 columns 2nd position
plt.hist(x = 'Defense', data = my_data, bins = 10, color = 'darkblue')
plt.title('Defense')
plt.xticks(np.arange(0,201,10), rotation = 45)
plt.yticks(np.arange(0,41,5))

plt.subplot(133)  ## 1 row 3 coulumn 3rd position
plt.hist(x = 'Speed', data = my_data, bins = 10, color = 'darkblue')
plt.title('Speed')
plt.show()
```

```python
#countplot: count plot is useful for visualizing the frequency distribution of discrete /categorical variable.

sns.countplot(x = 'Type 1', data = my_data, palette = 'Set1')

#rotate x labels

plt.xticks(rotation=45)
```

```python
sns.countplot(x = 'Type 1', data = my_data, palette = 'Set1', order= my_data['Type 1'].value_counts().sort_values(ascending= True).index)
plt.xticks(rotation=45)
```

```python
sns.countplot(x = 'Type 1', data = my_data, palette = 'Set1', order= my_data['Type 1'].value_counts().index.sort_values(ascending= True))
plt.xticks(rotation=45)
```

```python
# Bar plot: Bar graph is used to visualize the relationship between one categorical and one numerical variable.

sns.barplot(x = 'Type 1', y = 'Attack', data = my_data)

plt.xticks(rotation=90)
```

```python
#Boxplot :- is used to visualize the distribution of simple continuous variable but more widely used to 
#detect the outliers in the variable.
#Generate boxplots only for selected or individual variables for better representation.

sns.boxplot(data = my_data)
```

```python
sns.boxplot(my_data['HP'])
```

```python
#Line plot :- Is used to visualize a continuous variable over a period of time. e.g. stock price prediction, ECG

x = [10, 30, 20, 70]
y = [10, 45, 20, 70]
plt.plot(x,y,c = 'purple')

plt.xlim(0,100)
plt.ylim(0,100)
plt.title('A Simple Line Plot')
plt.grid(True)
plt.savefig('test2.jpg')
for a,b in zip(x,y):
    plt.text(a,b,(a,b))
plt.show()
```

```python
# Piechart :- Piecharts are useful to visualise the % distribution of a single categorical variable.
import matplotlib.pyplot as plt
# Slices will be plotted counter-clockwise:
labels = ['SQL','R','Tableau','Python']
sizes = [75,156,40,50]
explode = [0.2, 0, 0.1, 0]

plt.pie(sizes, explode = explode, labels= labels, autopct = "%1.2f%%",
       colors = ['red','purple','yellow','green'],
       shadow = True, startangle = 90,counterclock = True)
plt.show()
```
